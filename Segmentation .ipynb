```python
Brownspot = cv2.imread('/content/drive/MyDrive/Paddy leaf dataset /K-means Algorithm/CleanData2.jpg')
plt.title('Brownspot',size=15,color='blue')
plt.imshow(Brownspot)
plt.show()
Blast = cv2.imread('/content/drive/MyDrive/Paddy leaf dataset /K-means Algorithm/blast1.jpg')
plt.title('Leaf Blast',size=15,color='blue')
plt.imshow(Blast)
plt.show()
Blight = cv2.imread('/content/drive/MyDrive/Paddy leaf dataset /K-means Algorithm/blight2.jpg')
plt.title('Leaf Blight',size=15,color='blue')
plt.imshow(Blight)
plt.show()
Healthy = cv2.imread('/content/drive/MyDrive/Paddy leaf dataset /K-means Algorithm/healthydata.jpg')
plt.title('Healthy',size=15,color='blue')
plt.imshow(Healthy)
plt.show()


```

```python
Brownspot = cv2.cvtColor(Brownspot, cv2.COLOR_BGR2RGB)
plt.imshow(Brownspot)
plt.title('Brownspot',size=15,color='blue')
plt.show()

```

```python
hsv_Brownspot = cv2.cvtColor(Brownspot, cv2.COLOR_RGB2HSV)
plt.title('Brownspot',size=15,color='blue')
plt.imshow(hsv_Brownspot)


```

```python
hsv_Blast = cv2.cvtColor(Blast, cv2.COLOR_RGB2HSV)
plt.title('Leaf Blast',size=15,color='blue')
plt.imshow(hsv_Blast)
```

```python
lower_green1 = np.array([30,0,25]) #np array[lower mane kom green ta detect kore vanish kore fele,kom green mane green bad e baki ta detect korbe]
upper_green1 = np.array([100,255,255])# besi green detect kore
mask1 = cv2.inRange(hsv_Brownspot, lower_green1, upper_green1)
result1 = cv2.bitwise_and(Brownspot, Brownspot, mask=mask1)

plt.figure(figsize=(15,15))

plt.subplot(1, 2, 1)
plt.imshow(mask1, cmap="gray")
plt.title('Brownspot',size=15,color='blue')
plt.subplot(1, 2, 2)
plt.imshow(result1)
plt.title('Brownspot',size=15,color='blue')
plt.show()
```
